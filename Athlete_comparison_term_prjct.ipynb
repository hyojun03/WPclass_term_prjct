## 가설 3. 남자선수의 수가 여자선수의 수보다 많은 종목이 그 반대의 경우 보다 많을 것이다.
###### 데이터 가공 단계에서 만든 남자선수의 수 list와 여자선수의 수 list를 index를 활용하여 비교한다.
###### 이때 남자선수의 수가 여자선수의 수보다 많을 때는 man_win이라는 변수에 1을 더해주고, 그 반대일 경우에는 woman_win이라는 변수에 1을 더해준다. 두 값이 같을 경우에는 samesame이라는 변수에 1을 더해준다. 이를 통해 모든 경우를 구해본다

```python
man_win = 0 # 남자선수의 수가 여자선수의 수보다 많은 경우
woman_win = 0 # 남자선수의 수가 여자선수의 수보다 적은 경우
samesame = 0 # 남자선수의 수와 여자선수의 수가 같은 경우

for k in range(0,72):
    if df_man[k] == df_woman[k]:
        samesame += 1
    elif df_man[k] > df_woman[k]:
        man_win += 1
    else:
        woman_win += 1
        
print(man_win, woman_win,samesame)

col = ["남자선수가 많은 경우","여자선수가 많은 경우","같은 경우"]
ind = ["계"]
con = [man_win, woman_win, samesame]
df_whowin = pd.DataFrame(con,col,ind)
df_whowin
```

데이터를 가지고 각 지역의 모든 종목의 선수의 수를 구해본 결과, 서울 33568명, 부산 11253명, 인천 11965명, 대구 7857명, 대전 5972명, 광주 7784명, 울산 6252명이 나왔다. 

따라서 가설대로 7개의 지역중에서 서울지역의 선수의 수가 가장많았다. 

### 가설 3. 남자선수의 수가 여자선수의 수보다 많은 종목이 그 반대의 경우 보다 많을 것이다.

데이터를 가지고 남자선수의 수와 여자선수의 수를 종목별로 모두 비교해본 결과, 남자선수의 수가 여자선수의 수보다 많은 종목은 63개가 있었고, 그 반대의 경우는 8개, 같은 경우는 1개가 있었다.

따라서 가설대로 남자선수의 수가 여자선수의 수보다 많은 종목이 그 반대의 경우 보다 많았다.

### 가설 4.  전체 선수중에서 가장 많은 선수를 가진 집단은 일반부일 것이다.

데이터를 가지고 13세이하부, 16세이하부, 19세이하부, 대학부, 일반부로 나눠서 비교해본 결과, 13세이하부 36360명, 16세이하부 37323명, 19세이하부 29247명, 대학부 14729명, 일반부 160767명으로 일반부가 압도적으로 많았다.

또한 가설 1에서 축구선수의 수가 압도적으로 많은 이유를 분석하기 위하여 각 집단의 종목별 선수의 수를 원그래프로 나타내어보니 축구(60)선수의 수는 매 집단마다 20%이상의 비율을 차지하였고, 특히 그중 선수가 가장 많은 일반부에서는 62%를 차지하였다.

따라서 가설대로 전체 선수중에서 가장 많은 선수를 가진 집단은 일반부였다.
